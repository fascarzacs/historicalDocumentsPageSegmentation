```python
import cv2
import numpy as np
import os
from skimage.util import img_as_float
from skimage import io
import util
from timeit import default_timer as timer
from datetime import datetime
```

```python
folderDataset = "unetParzivalData"
xTrain = []; yTrain = []; yImgTrain = []
xTest = []; yTest = []; yImgTest = []
xValidation = []; yValidation = []; yImgValidation = []

```

```python
def readUnetImages(folder, subfolder) : 
    print(folderDataset + "/" + folder + "/" + subfolder)
    listImages = []
    for file in os.listdir(folderDataset + "/" + folder + "/" + subfolder) :
        if file.endswith('.jpg') :
            image = img_as_float(io.imread(folderDataset + "/" + folder + "/" + subfolder + "/" + file))
            listImages.append(image)
    return listImages
```

```python
def readUnetLabels(folder, subfolder) :
    print(folderDataset + "/" + folder + "/" + subfolder)
    listLabels = []
    for file in os.listdir(folderDataset + "/" + folder + "/" + subfolder) :
        if file.endswith('.npy') :
            labelNpArray = np.load(folderDataset + "/" + folder + "/" + subfolder + "/" + file)
            listLabels.append(labelNpArray)
    return listLabels
```

```python
xTrain = readUnetImages("train", "image")
xTest = readUnetImages("test", "image")
xValidation = readUnetImages("validation", "image")

yImgTrain      = readUnetImages("train", "label")
yImgTest       = readUnetImages("test", "label")
yImgValidation = readUnetImages("validation", "label")

yTrain = readUnetLabels("train", "label")
yTest = readUnetLabels("test", "label")
yValidation = readUnetLabels("validation", "label")
```
